### –¢–∞–±–ª–∏—Ü–∞ –¥–ª—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏:

```python
analisys = pd.DataFrame({'model':[], 'MAE_model':[], '–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã':[]})
```

### –ü–æ–¥–±–æ—Ä –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤:

```python
def train_model(model, param):
    
    model_gscv = GridSearchCV(estimator=model,
                                param_grid=param, 
                                scoring='neg_mean_absolute_error', 
                                cv=5, 
                                verbose=1)
    
    model_gscv.fit(features_train, target_train)
    
    
    mae = -1 * model_gscv.best_score_
    
    print('//n –õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:', model_gscv.best_params_)
    print('MAE –æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏:', round(mae, 2))
    
    row = []
    row.extend([model, mae, model_gscv.best_params_])
    analisys.loc[len(analisys.index)] = row
    
    return model_gscv
```

### –î–µ—Ä–µ–≤–æ —Ä–µ—à–µ–Ω–∏–π:

```python
tree_params = {"max_depth": list(range(3, 12))}

model_forest = DecisionTreeRegressor(random_state=state)

tree = train_model(model_forest, tree_params)
```

### –°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å:

```python
%time

forest_params = {
    'max_depth': list(range(8, 11)),
    'n_estimators': list(range(90, 100, 10))
}

model_forest = RandomForestRegressor(random_state=state, 
                                     n_jobs=-1)

forest = train_model(model_forest, forest_params)
```

### LightGBM:

```python
%time

lgbm_params = {
    'n_estimators': list(range(40, 50, 5)),
    'max_depth': list(range(5, 7)),
    'learning_rate': [0.001, 0.01, 0.1]
}

model_lgbm = lgb.LGBMRegressor(random_state=state,
                        boosting_type='gbdt',
                         objective = 'regression',
                         metric = 'RMSE')

lgbm = train_model(model_lgbm, lgbm_params)
```

### CatBoost:

```python
%time

cb_params = {
    'n_estimators': list(range(148, 152, 2)),
    'max_depth': list(range(5, 6)),
    'learning_rate': [0.001, 0.01, 0.1]
}

model_cb = CatBoostRegressor(verbose=False) 

cb = train_model(model_cb, cb_params)
```
