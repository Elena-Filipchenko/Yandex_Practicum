### –ü–æ–¥–±–æ—Ä –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤:

```python
def train_model(model, param):
    
    model_gscv = GridSearchCV(estimator=model,
                                param_grid=param, 
                                scoring='neg_mean_absolute_error', 
                                cv=5, 
                                verbose=1)
    
    model_gscv.fit(features_train, target_train)
    
    
    mae = -1 * model_gscv.best_score_
    
    print('//n –õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:', model_gscv.best_params_)
    print('MAE –æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏:', mae)
    
    row = []
    row.extend([model, mae, model_gscv.best_params_])
    analisys.loc[len(analisys.index)] = row
    
    return model_gscv
```

### –°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å:
