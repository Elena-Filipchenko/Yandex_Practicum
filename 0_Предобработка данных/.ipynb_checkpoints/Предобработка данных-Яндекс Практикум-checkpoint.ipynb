**Задание 5. В двух столбцах есть пропущенные значения. Один из них — `days_employed`. Пропуски в этом столбце вы обработаете на следующем этапе. Другой столбец с пропущенными значениями — `total_income` — хранит данные о доходах. На сумму дохода сильнее всего влияет тип занятости, поэтому заполнить пропуски в этом столбце нужно медианным значением по каждому типу из столбца `income_type`. Например, у человека с типом занятости `сотрудник` пропуск в столбце `total_income` должен быть заполнен медианным доходом среди всех записей с тем же типом.**

```python
for t in data['income_type'].unique():
    data.loc[(data['income_type'] == t) & (data['total_income'].isna()), 'total_income'] = \
    data.loc[(data['income_type'] == t), 'total_income'].median()
```

### Категоризация данных

**Задание 16. На основании диапазонов, указанных ниже, создайте в датафрейме `data` столбец `total_income_category` с категориями:**

- 0–30000 — `'E'`;
- 30001–50000 — `'D'`;
- 50001–200000 — `'C'`;
- 200001–1000000 — `'B'`;
- 1000001 и выше — `'A'`.


**Например, кредитополучателю с доходом 25000 нужно назначить категорию `'E'`, а клиенту, получающему 235000, — `'B'`. Используйте собственную функцию с именем `categorize_income()` и метод `apply()`.**

```python
def categorize_income(income):
    try:
        if 0 <= income <= 30000:
            return 'E'
        elif 30001 <= income <= 50000:
            return 'D'
        elif 50001 <= income <= 200000:
            return 'C'
        elif 200001 <= income <= 1000000:
            return 'B'
        elif income >= 1000001:
            return 'A'
    except:
        pass
```

```python
data['total_income_category'] = data['total_income'].apply(categorize_income)
```

**Задание 18. Создайте функцию, которая на основании данных из столбца `purpose` сформирует новый столбец `purpose_category`, в который войдут следующие категории:**

- `'операции с автомобилем'`,
- `'операции с недвижимостью'`,
- `'проведение свадьбы'`,
- `'получение образования'`.

**Например, если в столбце `purpose` находится подстрока `'на покупку автомобиля'`, то в столбце `purpose_category` должна появиться строка `'операции с автомобилем'`.**

**Используйте собственную функцию с именем `categorize_purpose()` и метод `apply()`. Изучите данные в столбце `purpose` и определите, какие подстроки помогут вам правильно определить категорию.**

```python
def categorize_purpose(row):
    try:
        if 'автом' in row:
            return 'операции с автомобилем'
        elif 'жил' in row or 'недвиж' in row:
            return 'операции с недвижимостью'
        elif 'свад' in row:
            return 'проведение свадьбы'
        elif 'образов' in row:
            return 'получение образования'
    except:
        return 'нет категории'
```

```python
data['purpose_category'] = data['purpose'].apply(categorize_purpose)
```

#### 3.1 Есть ли зависимость между количеством детей и возвратом кредита в срок?

Столбец 'children' указывает нам на уникальные значения количества детей в семье. Ранее мы скорректировали таблицу, удалив все аномальные значения. Таким образом, сейчас в таблице остались значения от 0 до 5 ([1, 0, 3, 2, 4, 5]). Сгруппируем данные в столбце 'children' по данным в столбце 'debt'  и посмотрим на результаты.
С учетом того, что за наличие просрочек по кредиту в столбце отвечает 1, а за их отсутствие 0, то с помощью методов count() и sum(), можно найти общее количество значений по каждому количеству детей и количество просрочек соответственно.

```python
children_debt = data.groupby('children')['debt'].count()    #общее количество по количеству детей
print(children_debt)
children_debt_delay = data.groupby('children')['debt'].sum()     #количество просрочек по каждому количеству детей
children_debt_delay 
```

Рассчитаем количество просрочек в процентном соотношении от общих значений по каждому варианту количества детей.

```python
children_debt_delay_percent = children_debt_delay / children_debt
children_debt_delay_percent
```

По полученным значениям видно, что частота просрочек лежит в диапазоне от 0 до 9,76%. Минимальное значение - 0% соответствует наличию в семье 5 детей, но общее количество таких кредитов - 9, что в случае данной таблицы является нечтожно малым и не может говорить о том, что выдача кредитов семьям с 5 детьми является стратегическо верной, что бы просрочек не было.

Так же отдельно стоит отметить, что чем меньше количество данных, тем менее на них можно правильно полагаться для утверждения результатов. Тем не менее, остальные значения лежат в более или менее одной области (от 7,54 до 9,76%, что соответствует 8,65 +- 1,11%). Данное значение лежит в пределах 13% отклонения от среднего, что является нормальным для данного количества данных. Однако, значение 13% отклонения может быть высоким для банковской сферы и стоит уточнить у заказчика, что является приемлемым. Это поможет сделать правильные выводы для конкретной сферы.

**Вывод:** В ходе анализа данных было выявлено, что при данном количестве данных количество детей не влияет на частоту просрочек по кредиту.

Чаще всего просрочки бывают у людей, которые берут кредит на авто (9,35%) и получают образование(9,25%).
